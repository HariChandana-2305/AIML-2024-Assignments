```python
import cv2
import os

# Define the list of labels
labels = ['Elephant', 'Zebra']

# Define the image size for resizing
img_size = 200

# Define the function to load training data
def get_train_data(data_dir):
    data = []
    for label in labels:
        path = os.path.join(data_dir, label)
        if not os.path.exists(path):
            print(f"Directory '{path}' does not exist.")
            continue
        class_num = labels.index(label)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path, img))
                img_arr = cv2.cvtColor(img_arr, cv2.COLOR_BGR2GRAY)  # Convert BGR to grayscale
                resized_arr = cv2.resize(img_arr, (img_size, img_size))
                data.append([resized_arr, class_num])
            except Exception as e:
                print(f"Error processing '{os.path.join(path, img)}': {e}")
    return data
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate synthetic dataset with adjusted parameters
X, y = make_classification(n_samples=150, n_features=4, n_informative=2, n_classes=2, n_clusters_per_class=2, random_state=42)

# Define the number of bootstrap iterations
n_bootstrap_iterations = 1000
# Initialize an array to store accuracy scores
accuracy_scores = np.zeros(n_bootstrap_iterations)

# Define the logistic regression classifier
log_reg = LogisticRegression(max_iter=1000)  # Adjust max_iter as needed

# Perform bootstrap iterations
for i in range(n_bootstrap_iterations):
    # Create bootstrap sample indices
    bootstrap_indices = np.random.choice(len(X), size=len(X), replace=True)
    # Create bootstrap samples
    X_bootstrap = X[bootstrap_indices]
    y_bootstrap = y[bootstrap_indices]

    # Fit logistic regression classifier
    log_reg.fit(X_bootstrap, y_bootstrap)

    # Predict on the entire dataset
    y_pred = log_reg.predict(X)

    # Calculate accuracy
    accuracy_scores[i] = accuracy_score(y, y_pred)

# Calculate mean accuracy
mean_accuracy = np.mean(accuracy_scores)
print("Mean Accuracy:", mean_accuracy)

# Calculate standard deviation of accuracy
std_accuracy = np.std(accuracy_scores)
print("Standard Deviation of Accuracy:", std_accuracy)

# Calculate 95% confidence interval
confidence_level = 0.95
confidence_interval = (1 - confidence_level) / 2
lower_bound = np.percentile(accuracy_scores, confidence_interval * 100)
upper_bound = np.percentile(accuracy_scores, (1 - confidence_interval) * 100)
print("95% Confidence Interval: [{:.4f}, {:.4f}]".format(lower_bound, upper_bound))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have already computed these variables for Logistic Regression
sample_numbers = np.arange(1, n_bootstrap_iterations + 1)
logistic_regression_bootstrap_accuracies = accuracy_scores
logistic_regression_mean_accuracy = np.mean(accuracy_scores)
logistic_regression_std_accuracy = np.std(accuracy_scores)
confidence_level = 0.95
logistic_regression_conf_interval = np.percentile(accuracy_scores, [100 * (1 - confidence_level) / 2, 100 * (1 + confidence_level) / 2])

plt.figure(figsize=(10, 6))
plt.plot(sample_numbers, logistic_regression_bootstrap_accuracies, label="Bootstrap Accuracy", color='r')
plt.axhline(logistic_regression_mean_accuracy, linestyle='--', color='b', label="Mean Accuracy")

plt.fill_between(sample_numbers, logistic_regression_conf_interval[0], logistic_regression_conf_interval[1], alpha=0.3, color='c', label="95% CI")

plt.xlabel("Bootstrap Sample")
plt.ylabel("Accuracy")
plt.title("Bootstrap Method on Logistic Regression Model")
plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification
from scipy.stats import sem, t

# Generate synthetic dataset
X, y = make_classification(n_samples=150, n_features=4, n_classes=2, n_clusters_per_class=2, n_informative=2, random_state=42)

# Define x_train and y_train
x_train = X
y_train = y

# Check the dimensions of x_train and y_train
print("Dimensions of x_train:", x_train.shape)
print("Dimensions of y_train:", y_train.shape)

# Ensure that the length of y_train matches the number of samples in x_train
if len(y_train) != len(x_train):
    print("Error: Number of labels does not match number of samples in the training data.")
else:
    # Define the number of bootstrap iterations
    n_bootstrap_iterations = 1000
    # Initialize an array to store accuracy scores
    accuracy_scores = np.zeros(n_bootstrap_iterations)

    # Define the Decision Tree classifier
    decision_tree = DecisionTreeClassifier()

    # Fit the Decision Tree classifier to the training data
    decision_model = decision_tree.fit(x_train, y_train)

    # Perform bootstrap iterations
    for i in range(n_bootstrap_iterations):
        # Create bootstrap sample indices
        bootstrap_indices = np.random.choice(len(x_train), size=len(x_train), replace=True)
        # Create bootstrap samples
        X_bootstrap = x_train[bootstrap_indices]
        y_bootstrap = y_train[bootstrap_indices]

        # Fit Decision Tree classifier
        decision_tree.fit(X_bootstrap, y_bootstrap)

        # Predict on the entire dataset
        y_pred = decision_tree.predict(X_bootstrap)  # Modified to predict on the bootstrap sample

        # Calculate accuracy
        accuracy_scores[i] = accuracy_score(y_bootstrap, y_pred)  # Modified to calculate accuracy on bootstrap sample

    # Calculate mean accuracy
    mean_accuracy = np.mean(accuracy_scores)
    print("Mean Accuracy:", mean_accuracy)

    # Calculate standard deviation of accuracy
    std_accuracy = np.std(accuracy_scores)
    print("Standard Deviation of Accuracy:", std_accuracy)

    # Calculate 95% confidence interval
    confidence_level = 0.95
    confidence_interval = sem(accuracy_scores) * t.ppf((1 + confidence_level) / 2, len(accuracy_scores) - 1)
    print("95% Confidence Interval:", confidence_interval)
```
